```python
from numpy import *

def f1(x, xm):
  return 1/(sin(xm/2)**2-sin(x/2)**2)**(1/2)

def T(g,x1,x2): #Integración por el metodo del trapecio
  M=100
  h=(x2-x1)/M
  s=0.5*g(x1,x2)+0.5*g(x2-0.001,x2)        # se resta 0.001 al limite superior para evitar dividir entre 0 y para mantenerse en el intervalo que nos interesa
  for k in range(1,M):
    s+=g(x1+(k*h),x2)
  return h*s

def S(g,x1, x2): # Integración por el metodo de Simpson
  M=100
  h=(x2-x1)/(M)
  s= g(x1,x2)+g(x2-0.001,x2)
  for k in range(1,M):
    if k%2==0:
      s+=2*g(x1+k*h,x2)
    else:
      s+=4*g(x1+k*h,x2)
  return (1/3)*h*s

def f2(x):
  return 1+ (1/2)**2 * sin(x/2)**2 +(3/(2*4))**2 * sin(x/2)**4

L=pi/50 # tamaño del intervalo para evaluar la función
y=L*arange(1,50) # Dominio de la función resultante de la integral (el arange generara numeros desde el 1 al 49)

I1=T(f1,0.001,y)/pi       #Integral calculada por el metodo del trapecio
                  # array de la integral evaluada en los limites determinados por 'y'

I2=S(f1,0.001,y)/pi    # Integral calculada por el metodo de Simpson


I3=f2(y)
```

```python
import matplotlib.pyplot as plt

plt.plot(y,I1, 'm', label="M.Del trapecio")
plt.plot(y,I2, 'r', label="Simpson")
plt.plot(y,I3, 'b', label= "Serie Truncada")
plt.legend()

plt.xlabel('$\Theta_m$', fontsize=20)
plt.ylabel('T/To', fontsize=20)
```
