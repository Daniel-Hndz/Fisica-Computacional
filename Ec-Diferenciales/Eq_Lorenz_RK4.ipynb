```python
from numpy import *
def f(r,t):  # 'r' es el vector que contiene las variables
  p=10.0
  u=28.0
  b=8/3
  x,y,z=r[0],r[1], r[2]
  fx=p*(y-x)
  fy=(u*x)-y-(x*z)
  fz=(x*y)-(b*z)
  return array([fx,fy,fz],float)

a,b=0,50
N=10000
h=(b-a)/N

tpts=arange(a,b,h) # dominio (tiempo)

xpts=zeros(len(tpts)) #coordenadas
ypts=copy(xpts)
zpts=copy(xpts)

r=array([1,0,0],float) # Valores iniciales, [x_0, y_0, z_0]
i=0
for t in tpts:
  xpts[i],ypts[i],zpts[i]=r[0],r[1],r[2]  #creación de los puntos a graficar
  k1=h*f(r,t)
  k2=h*f(r+0.5*k1,t+0.5*h)
  r+=k2
  i+=1
```

```python
import matplotlib.pyplot as plt

plt.title('Coordenada y en función del tiempo')
plt.plot(tpts, ypts, 'g')
plt.xlabel('$t\ (s)$')
plt.ylabel('$y$')
plt.show()
print()

plt.title('Atractor extraño')
plt.plot(xpts, zpts, color='purple')
plt.xlabel('$x$')
plt.ylabel('$z$')
plt.show()
print()
```
