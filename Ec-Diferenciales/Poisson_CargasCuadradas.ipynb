```python
from numpy import *
from numpy import max
import matplotlib.pyplot as plt

def dens():
  rho=zeros([101,101],float)
  for i in range(101):
    for j in range(101):
      if (i>=20 and i<=40) and (j>=20 and j<=40):
        rho[i,j]=-1
      elif (i>=60 and i<=80) and (j>=60 and j<=80):
        rho[i,j]=1
  return rho

a=1
M=100
e=1e-2
eps=1#8.854e-12

phi=zeros([M+1,M+1],float)
phi[0,:]=0
phi[:,0]=0
phi[M,:]=0
phi[:,M]=0
phiprime=zeros([M+1,M+1],float)
rho=dens()
delta=1.0
while delta>e:
  for i in range(M+1):
    for j in range(M+1):
      if i==0 or j==0 or i==M or j==M:
        phiprime[i,j]=phi[i,j]
      else:
        phiprime[i,j]=(1/4)*(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1]+(a**2/eps)*rho[i,j])

  delta=max(abs(phi-phiprime))
  phi,phiprime=phiprime,phi

plt.imshow(phi, origin= 'lower' ,cmap='seismic', vmin=-200, vmax=200)
plt.title('Potencial Electrostático', fontsize=20)

plt.xlabel('x (cm)', fontsize=15)
plt.ylabel('y (cm)',fontsize=15)
plt.show()
```
