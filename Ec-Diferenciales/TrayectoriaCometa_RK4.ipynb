```python
from numpy import *

def f(r,t):  # 'r' es el vector que contiene las variables, tanto de posición como de velocidad
  x,y=r[0],r[1] # r=[x, y, v_x, v_y]
  v0,v1=r[2],r[3]
  fr=array([v0,v1,-x/sqrt((x**2)+(y**2))**3,
            -y/sqrt((x**2)+(y**2))**3], float) # fr= [fx, fy, fvx, fvy]
  return fr

a,b=0,210
N=10000
h=(b-a)/N

tpts=arange(0,210,h) # dominio (tiempo)

xpts=zeros(len(tpts)) #coordenadas
ypts=copy(xpts)

vpts=copy(xpts) #velocidades
upts=copy(xpts)

r=array([0.5,0,0,1.65],float) # Valores iniciales, [x_0, y_0, v_x0, v_y0]
i=0
for t in tpts:
  xpts[i],ypts[i],vpts[i],upts[i]=r[0],r[1],r[2],r[3] #creación de los puntos a graficar
  k1=h*f(r,t)
  k2=h*f(r+0.5*k1,t+0.5*h)
  k3=h*f(r+0.5*k2,t+0.5*h)
  k4=h*f(r+k3,t+h)
  r+=(k1+2*k2+2*k3+k4)/6
  i+=1
```

```python
import matplotlib.pyplot as plt
#Grafica de trayectoria
plt.title('Trayectoria del cometa')
plt.plot(xpts, ypts, 'g')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
print()

rap=sqrt(vpts**2+upts**2)   # rapidez del cometa

#Grafica de la rapidez
plt.title('Rapidez del cometa')
plt.plot(tpts,rap ,'r')
plt.xlabel('$t $')
plt.ylabel('$v(t)$')
plt.show()
print()

#Grafica extra, con el objetivo de que se aprecie mejor como varia la rapidez en un ciclo aproximado
plt.title('Rapidez del cometa (un ciclo)')
plt.plot(tpts,rap , color='purple')
plt.xlim((0,8.71))
plt.xlabel('$t $')
plt.ylabel('$v(t)$')
plt.show()
```
