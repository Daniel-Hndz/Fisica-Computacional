
Problema de potencial de pozo cuadrado



```python
from numpy import *
#from scipy.constants import h

hbarc=0.1973 #eV*micrometros (constante reducida de planck por la constante c)
m=0.511e6 #eV/c^2
V=20 #eV
w=1e-3 #micrometros

def f1(E):
  return tan(sqrt((w**2*m*E)/(2*hbarc**2)))

def f2(E):
  return sqrt((V-E)/E)

def f3(E):
  return -sqrt(E/(V-E))


a=arange(0.1,V,0.1)

import matplotlib.pyplot as plt


plt.plot(a,f1(a))
plt.plot(a,f2(a))
plt.plot(a,f3(a))
plt.ylim(-15,15)
```

:Los primeros 5 estados excitados se encuentran en los rangos (No se toman en cuenta las lineas verticales):
$$1\leq E_1 \leq 1.5$$
$$2.5\leq E_2 \leq 3$$
$$4\leq E_3 \leq 5.5$$
$$7\leq E_4 \leq 8$$
$$10\leq E_5 \leq 12 $$

```python

def bisection (xminus, xplus, f):
  eps=0.0001    #parametro de precisión
  Nmax=1000
  j=0
  x=xplus
  while abs(f(x))> eps  and j < Nmax :
    x=(xplus + xminus)/2.0                             #Punto medio

    if (f(xplus)*f(x) > 0.0 ) :
      xplus=x
    else:
      xminus=x
    j+=1
  if j == Nmax:
    print ( f" Raíz no encontrada despues de {Nmax} iteraciones \n" )
  return x

def p(E): #Estados pares
  return f1(E)-f2(E)

def i(E): #Estados impares
  return f1(E)-f3(E)

E1=bisection(1,1.5,i)
E2=bisection(2.5,3,p)
E3=bisection(4,5.5,i)
E4=bisection(7,8,p)
E5=bisection(10,12,i)

print(f"E1= {E1:.3f} eV\n E2= {E2:.3f}eV\n E3= {E3:.3f} eV\n E4= {E4:.3f}eV\n E5= {E5:.3f}eV ")
```
