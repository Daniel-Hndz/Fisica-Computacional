```python
from numpy import *
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

a1,a2=0,2        #limites de la grafica en el eje x
b1,b2=-1.5,1.5   #limites de la grafica en el eje y

def f(x,y):
  return x**2-2*x+y**4-2*y**2+y

fig=plt.figure()
ax = plt.axes(projection='3d')

X=linspace(a1,a2,50)
Y=linspace(b1,b2,50)
X,Y=meshgrid(X,Y)

ax.contour3D(X, Y, f(X,Y), 100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax.view_init(10,-20)
fig
```

Se pueden observar dos minimos en las cercanias de [x,y]=(1,-1) y (1,1)
